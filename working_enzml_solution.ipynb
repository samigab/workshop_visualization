## Section 4
### Michaelis-Menten Model Fitting and Visualization

This section demonstrates how to define and fit a **Michaelis-Menten enzyme kinetics model** to experimental data, and then plot the fitted curve.

---

#### Step 1: We define the Michaelis-Menten Equation
#### Step 2: Fit the Model to Data

```python
popt, _ = curve_fit(michaelis_menten, data_mm["S"], data_mm["v0"], p0=[1, 10])
```

* `curve_fit` is a SciPy function that performs non-linear least squares optimization to fit the model to data.
* **Inputs**:

  * `michaelis_menten`: The function we want to fit
  * `data_mm["S"]`: Substrate concentration data
  * `data_mm["v0"]`: Initial reaction rate data
  * `p0=[1, 10]`: Initial guesses for `[Vmax, Km]`
* **Outputs**:

  * `popt`: The optimized parameters `[Vmax, Km]` that best fit the data

---

#### Step 3: Generate Fitted Curve
```python
S_fit = np.linspace(0, 160, 100) #
v_fit = michaelis_menten(S_fit, *popt)
```

* `S_fit`: A smooth range of substrate concentrations to plot the fitted curve
* `linspace`: creates 100 equally spaced points between 0 and 160, the borders of our measurement data
* `v_fit`: The corresponding reaction rates computed from the fitted model

---

#### Step 4: Plot the Fitted Model

```python
plt.plot(S_fit, v_fit, label=f"Fit: Vmax={popt[0]:.2f}, Km={popt[1]:.2f}")
```

* This draws the fitted Michaelis-Menten curve
* The `label` shows the fitted parameter values (rounded to two decimal places)
* You can add this to a plot with your experimental data to compare fit vs. real measurements


````python
#Step 1:
def michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

#Step 2:
popt, _ = curve_fit(michaelis_menten, data_mm["S"], data_mm["v0"], p0=[1, 10])

#Step 3:
S_fit = np.linspace(0, 160, 100)
v_fit = michaelis_menten(S_fit, *popt)

plt.plot(S_fit, v_fit, label=f"Fit: Vmax={popt[0]:.2f}, Km={popt[1]:.2f}")
### 
### Add your Scatterplot from above!
###
plt.legend()
plt.show()
````

## Section 5
### Subplots Multible Plots in One!

This section demonstrates how to apply three classic **linear transformations** to enzyme kinetics data to estimate kinetic parameters using linear regression.

These methods pre-date modern nonlinear fitting.
- Lineweaver-Burk transformation: Reciprocal plot
- Eadie-Hofstee transformation: v vs. v/[S]
-  Hanes-Woolf transformation: [S] vs. [S]/v

````python
#  Transform to Lineweaver-Burk
inv_S = 1 / data_mm["S"]
inv_v = 1 / data_mm["v0"]

#  Transform to # Eadie-Hofstee
EH_x = data_mm["v0"] / data_mm["S"]
EH_y = data_mm["v0"]

# Transform to Hanes-Woolf
HW_x = data_mm["S"] / data_mm["v0"]
HW_y = data_mm["S"]

# Set up 1x3 grid
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Lineweaver-Burk
axs[0].scatter(inv_S, inv_v, color='purple')
axs[0].set_title("Lineweaver-Burk Plot")
axs[0].set_xlabel("1/[S]")
axs[0].set_ylabel("1/v₀")

# Eadie-Hofstee
axs[1].scatter(EH_x, EH_y, color='green')
axs[1].set_title("Eadie-Hofstee Plot")
axs[1].set_xlabel("v₀/[S]")
axs[1].set_ylabel("v₀")

# Hanes-Woolf
axs[2].scatter(HW_x, HW_y, color='darkorange')
axs[2].set_title("Hanes-Woolf Plot")
axs[2].set_xlabel("[S]/v₀")
axs[2].set_ylabel("[S]")

#Show
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Recreate the simulated data if not already present
substrate_concs = [0.05, 0.10, 0.25, 0.50, 1.00, 2.50, 5.00, 8.00, 20.00, 30.00]
velocities = {
    "v3": [0.004, 0.008, 0.020, 0.035, 0.060, 0.110, 0.138, 0.154, 0.179, 0.200],
    "v4": [0.004, 0.007, 0.020, 0.035, 0.056, 0.104, 0.138, 0.150, 0.179, 0.200],
    "v5": [0.004, 0.007, 0.016, 0.032, 0.050, 0.090, 0.115, 0.119, 0.142, 0.166],
    "v6": [0.003, 0.006, 0.017, 0.031, 0.048, 0.101, 0.121, 0.139, 0.152, 0.181]
}

timepoints = np.arange(0, 9, 1)  # 0 to 8 minutes, 1-minute interval

# Simulate data with noise
data = []
np.random.seed(42)  # for reproducibility
for i, conc in enumerate(substrate_concs):
    for t in timepoints:
        row = {"ONPG": conc, "time": t}
        for v_name in velocities:
            v = velocities[v_name][i]
            noise = np.random.normal(0, 0.05)  # Add small Gaussian noise
            row[v_name] = max(0, v * t + noise)  # Ensure non-negative
        data.append(row)

sim_noisy_df = pd.DataFrame(data)

# Plotting
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True, sharey=True)
axes = axes.flatten()
velocity_cols = ['v3', 'v4', 'v5', 'v6']

for ax, v in zip(axes, velocity_cols):
    for conc in substrate_concs:
        subset = sim_noisy_df[sim_noisy_df['ONPG'] == conc]
        ax.plot(subset['time'], subset[v], marker='o', label=f'{conc} mM')
    ax.set_title(f'Product vs Time with Noise - {v}')
    ax.set_xlabel('Time (min)')
    ax.set_ylabel('Product Conc. (a.u.)')
    ax.legend(title='[ONPG] (mM)', fontsize='small')

plt.tight_layout()
plt.show()
````

````python
from sklearn.linear_model import LinearRegression


def get_velocities_from_df(pandas_df, species):
    """
    Perform linear regression on grouped data by 'id'
    to calculate initial velocities (slopes) of 's2' vs 'time'.
    
    Parameters:
        pandas_df (pd.DataFrame): DataFrame containing 'time', and 'id' columns.
        species (str): String of species name
    Returns:
        dict: A dictionary mapping each 'id' to the slope (velocity).
    """
    slopes = {}
    
    for name, group in pandas_df.groupby("id"):
        X = group["time"].values.reshape(-1, 1)
        y = group[species].values
        model = LinearRegression().fit(X, y)
        slopes[name] = model.coef_[0]
    
    return slopes

velocities = get_velocities_from_df(pandas_df, "s2")
# Print slopes
for id_, velocity in velocities.items():
    print(f"Slope for {id_}: {velocity:.6f} units/s")


concs = [0.05,0.1]
````

````python
def get_initial_data(enz_obj, species):
    """
    Extract and sort initial concentrations of species species from measurements.

    Parameters:
        enz_obj: An EnzymeMLDocument with a method `filter_measurements()` returning a list of Measurement objects.
        species. str of species name
    Returns:
        List[Tuple[str, float]]: Sorted list of (measurement_id, initial_concentration).
    """
    initial_data = []

    for m in enz_obj.filter_measurements():
        for sd in m.species_data:
            if sd.species_id == species:
                initial_data.append((m.id, sd.initial))

    initial_data_s = sorted(initial_data, key=lambda x: x[0])
    return initial_data_s


init_conc = get_initial_data(onpg_data, "s1")
# Print the results
for measurement_id, initial_value in init_conc:
    print(f"Measurement ID: {measurement_id}, Initial [s1]: {initial_value}")
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Michaelis-Menten function
def michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

# Time setup
dt_fine = 0.5
t_end = 240
time_fine = np.arange(0, t_end + dt_fine, dt_fine)
time_points = np.arange(0, t_end + 1, 15)

# "Nominal" enzyme parameters (will be jittered)
enzymes = {
    "Enzyme_A": {"Km": 30, "Vmax": 0.8},
    "Enzyme_B": {"Km": 50, "Vmax": 2.5},
    "Enzyme_C": {"Km": 20, "Vmax": 1.0},
    "Enzyme_D": {"Km": 40, "Vmax": 1.8},
}

# Substrate concentration baseline
S0_mean = 50.0
S0_std = 3.0  # more variability in starting concentration

# Jitter settings
km_jitter = 0.30   # ±10%
vmax_jitter = 0.30  # ±15%

# Simulation with jitter + more noise
def simulate_curve(time_fine, output_times, S0, Vmax, Km, noise_std=1.2):
    S = [S0]
    for i in range(1, len(time_fine)):
        rate = michaelis_menten(S[-1], Vmax, Km)
        new_S = S[-1] - rate * dt_fine
        S.append(max(new_S, 0))
    # Sample every 15 min + noise
    sampled = []
    for t in output_times:
        idx = int(t / dt_fine)
        noisy = S[idx] + np.random.normal(0, noise_std)
        sampled.append(max(noisy, 0))
    return sampled

# Generate data with more inter-curve diversity
results = []
np.random.seed(42)  # reproducibility
for enzyme, base_params in enzymes.items():
    # Jitter Km and Vmax
    Km = base_params["Km"] * (1 + np.random.uniform(-km_jitter, km_jitter))
    Vmax = base_params["Vmax"] * (1 + np.random.uniform(-vmax_jitter, vmax_jitter))
    S0 = np.random.normal(S0_mean, S0_std)
    
    S_values = simulate_curve(time_fine, time_points, S0, Vmax, Km)
    
    df = pd.DataFrame({
        "Time_min": time_points,
        "Substrate_uM": S_values,
        "Enzyme": enzyme
    })
    results.append(df)

# Combine all
full_data = pd.concat(results, ignore_index=True)

# Plot: scatter only
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=full_data, x="Time_min", y="Substrate_uM", hue="Enzyme", style="Enzyme", s=70)
plt.title("Diverse and Noisy Substrate Depletion Curves (15 min intervals)")
plt.ylabel("Substrate Concentration (µM)")
plt.xlabel("Time (minutes)")
plt.savefig("timecourse_.png", dpi=300)
plt.tight_layout()
plt.show()
````

````python
from scipy.integrate import solve_ivp

def michaelis_menten_ode(t, S, Vmax, Km):
    """
    Michaelis-Menten differential equation: dS/dt = -Vmax*S/(Km + S)
    S: substrate concentration
    Vmax: maximum reaction velocity 
    Km: Michaelis constant
    """
    return -Vmax * S / (Km + S)

# Prepare data for each enzyme separately
enzymes = full_data['Enzyme'].unique()
plt.figure(figsize=(12, 8))

for enzyme in enzymes:
    # Filter data for this enzyme
    enzyme_data = full_data[full_data['Enzyme'] == enzyme]
    t = enzyme_data['Time_min'].values
    S = enzyme_data['Substrate_uM'].values
    
    try:
        # Initial parameter guesses
        S0_guess = S[0]  # Initial substrate concentration
        Vmax_guess = (S[0] - S[-1]) / (t[-1] - t[0])  # Rough estimate
        Km_guess = S0_guess / 2  # Rough estimate
        
        def obj_function(params):
            Vmax, Km = params
            # Solve ODE
            sol = solve_ivp(
                lambda t, y: michaelis_menten_ode(t, y, Vmax, Km),
                [0, max(t)],
                [S0_guess],
                t_eval=t,
                method='RK45'
            )
            return np.sum((sol.y[0] - S)**2)
        
        # Optimize parameters using least squares
        from scipy.optimize import minimize
        result = minimize(
            obj_function, 
            [Vmax_guess, Km_guess],
            bounds=[(0, None), (0, None)]
        )
        
        Vmax_fit, Km_fit = result.x
        
        # Generate smooth curve for plotting
        t_smooth = np.linspace(0, max(t), 100)
        sol_smooth = solve_ivp(
            lambda t, y: michaelis_menten_ode(t, y, Vmax_fit, Km_fit),
            [0, max(t)],
            [S0_guess],
            t_eval=t_smooth,
            method='RK45'
        )
        
        # Plot data and fit
        plt.scatter(t, S, label=f'{enzyme} (data)')
        plt.plot(t_smooth, sol_smooth.y[0], '--',
                label=f'{enzyme} (fit)\nVmax={Vmax_fit:.2f}, Km={Km_fit:.2f}')
        
        print(f"\nResults for {enzyme}:")
        print(f"Vmax = {Vmax_fit:.2f}")
        print(f"Km = {Km_fit:.2f}")
        print(f"S0 = {S0_guess:.2f}")
        
    except RuntimeError:
        print(f"\nWarning: Could not fit {enzyme} data")
        continue

plt.xlabel('Time (minutes)')
plt.ylabel('Substrate Concentration (µM)')
plt.title('Michaelis-Menten Time Course Fits')
plt.legend()
plt.grid(True)
plt.savefig("timecourse_2.png", dpi=300)
plt.show()
````
